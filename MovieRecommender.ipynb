```python
# genres
# id
# keywords
# title
# overview
# cast
# crew

movies = movies[['id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.dropna(inplace=True)
```

```python
def convert(obj): #creating this func for genres
   L = []
   for i in ast.literal_eval(obj):
    L.append(i['name'])
   return L
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
   L=[]
   counter = 0
   for i in ast.literal_eval(obj):
      if counter != 3:
        L.append(i['name'])
        counter += 1
      else:
        break
   return L;
```

```python
print(credits.columns)
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
   L=[]
   for i in ast.literal_eval(obj):
    if i['job'] == 'Director' :
      L.append(i['name'])
      break
   return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview']= movies['overview'].apply(lambda x: x if isinstance(x, list) else x.split())
```

```python
movies.head()
```

```python
#to make "sam Worthington" into "samWorthington"...we will do this for all words
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords']+ movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['id','title','tags']]
```

```python
new_df
```

```python
#converting list into strings
new_df['tags']= new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
#converting all into lower case
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower() if isinstance(x, str) else " ".join(x).lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity[0]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title']==movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1]) [1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
